## Popularity Based Recommendation System Model Building

```python
books.head()
```

```python
ratings.head()
```

```python
# We merge both the Data Sets on the basis of 'ISBN' column
ratings_with_name = ratings.merge(books , on='ISBN')
```

```python
ratings_with_name.head()
```

```python
num_rating_df = ratings_with_name.groupby(['Book-Title']).count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating': 'Num_rating'} , inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby(['Book-Title']).mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating': 'Avg_rating'} , inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df , on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['Num_rating']>=250].sort_values('Avg_rating' , ascending = False).head(50)
```

```python
# This is the Popularity Based Recommendation System
mmmdf = popular_df.merge(books , on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Num_rating','Avg_rating','Image-URL-M']]
```

```python
mmmdf
```

## Collaborative Filtering Based Recommendation System

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
padhe_likhe_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
similarity_score[0]
```

```python
def recommend(book_name):
    index = np.where(pt.index ==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key = lambda x:x[1] ,reverse=True)[1:6]
    
    for i in similar_items: 
        print(pt.index[i[0]])
```

```python
recommend('1984')
```

```python
import pickle
pickle.dump(pt,open('pt.pkl','wb'))

pickle.dump(pt,open('ratings_with_name.pkl','wb'))
```
